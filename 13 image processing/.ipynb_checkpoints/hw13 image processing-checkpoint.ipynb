```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
import numpy as np
from skimage import io 

from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage.color import rgb2gray
from skimage import measure

from image_manipulation import prepare_landsat_image
```

```python
def PrintArrayInfo(a):
    print("Array info:")
    print("shape:", a.shape)
    print("dtype:", a.dtype)
    print("min, max:", a.min(), a.max())
```

## Part 1 - How much hair and beard?

```python
image = io.imread("head_by_Ron_Mueck.jpg")
plt.axis('off') # turn off axis labels
plt.imshow(image)
plt.show()
```

```python
gray_image = rgb2gray(image)
entropy_image = entropy(gray_image, disk(7)) 

plt.imshow(entropy_image)
plt.show()
```

```python
PrintArrayInfo(entropy_image)
```

```python
plt.hist(entropy_image.ravel(),100)
plt.show()
```

```python
val = 5.4

image_bw1 = (entropy_image >  val)
image_bw2 = (entropy_image <= val)

plt.imshow(image_bw1, cmap='gray')
plt.show()
```

```python
flattenedImg = image_bw1.flatten()
numPixels = len(flattenedImg)
numHair = len(flattenedImg[flattenedImg == True])
print(str(100 * numHair / numPixels) + "% is hair")
print(str(numHair / numPixels) + " is hair")
```

### For this image, we may overestimate the percentage of hair pixels because of the ear, background, some of the eyelid beinng picked up as being darker than the rest of the image. Therefore, these parts of the image will show up as white or True in our array, which we are using to indicate a hair pixel.

### We may underestimate as it's difficult to get all of the hair to turn white without getting other parts of the face that don't have hair to turn white. For example, the hair on the head doesn't have as many filled in white pixels as it should

## Part 2 - Bubbles in Pumice 2

```python
image = io.imread('3.jpg')
plt.axis('off') # turn off axis labels
plt.imshow(image, cmap = "gray")
plt.show()
```

```python
plt.hist(image.ravel(),100)
plt.show()
```

```python
val1 = 200
val0 = 100
image_grains = (image < val1 ) * (image > val0 )

plt.imshow(image_grains, cmap = "gray")
plt.show()
```

```python
labels = measure.label(image_grains)
print(np.unique(labels))
PrintArrayInfo(labels)
```

```python
props = measure.regionprops(labels)

# select the 8th region
grain = props[7]

# get the coordinates of the bubble
x1, y1, x2, y2 = grain.bbox

# Make a new image with the dimensions of the original one
img = np.zeros(image_grains.shape,dtype=bool)

# place the bubble image in the correct location in the new image
img[x1:x2,y1:y2] = grain.image
plt.imshow(img,cmap='gray')
plt.show()
```

```python
areas = []
for p in props:
    areas.append(p.area)
plt.hist(areas, 10)
plt.show()
```

```python
grainProps = props
```

### 295 grains detected

```python
image_bubbles = (image < 200)

plt.imshow(image_bubbles, cmap = "gray")
plt.show()
```

```python
labels = measure.label(image_bubbles)
print(np.unique(labels))
PrintArrayInfo(labels)
```

```python
props = measure.regionprops(labels)

# select the 8th region
bubble = props[7]# get the coordinates of the bubble
x1, y1, x2, y2 = bubble.bbox

# Make a new image with the dimensions of the original one
img = np.zeros(image_bubbles.shape,dtype=bool)

# place the bubble image in the correct location in the new image
img[x1:x2,y1:y2] = bubble.image
plt.imshow(img,cmap='gray')
plt.show()
```

### 599 bubbles detected

##### (4) Generate a histogram of areas for BOTH the bubbles and the grains. Once again, you can iterate through the area of each entry in the regionprops object using the starting with the following loop.

```python
areas = []
for p in props:
    areas.append(p.area)
plt.hist(areas, 10)
plt.show()
```

```python
bubbleProps = props
```

##### (5) Calculate the total area of bubbles and grains. Calculate the fraction of the image by area take up by the bubbles and the grains.

```python
grainArea = 0
for p in grainProps:
    grainArea += p.area
grainArea
```

```python
bubbleArea = 0
for p in bubbleProps:
    bubbleArea += p.area
bubbleArea
```

```python
totalArea = image_bubbles.shape[0] * image_bubbles.shape[1]
totalArea
```

```python
print("Percentage of grain area: " + str(grainArea / totalArea))
print("Percentage of bubble area: " + str(bubbleArea / totalArea))
```

## Part 3 - False Color LANDSAT Image

```python
from skimage import img_as_ubyte, img_as_uint, img_as_float64
from image_manipulation import prepare_landsat_image
bands = [6,5,2]
ls_false = prepare_landsat_image(bands)
```

```python
plt.imshow(ls_false)
plt.show()
```

```python
from PIL import Image

#Load Image into PIL
imageRGB = Image.fromarray(img_as_ubyte(ls_false))

N = 7
imageIndexed = imageRGB.quantize(colors=N, kmeans=1)
rgbFromIndexed = imageIndexed.convert("RGB")
```

```python
rgbColors = rgbFromIndexed.getcolors()
indexedColors = imageIndexed.getcolors()
counts = []
palette = []
labels = []

for color in rgbColors:
    count = color[0]
    counts.append(count)
    palette.append(color[1])
    
    for color in indexedColors:
        if color[0] == count:
            labels.append(color[1])
```

```python
from image_manipulation import bar_plot_with_colors
bar_plot_with_colors(counts,palette,labels)
```

#### beige = sparsely vegetated areas, light blue = urban areas, the muted and brighter greens are both vegetation (bright green = forest, muted green = stressed vegetation), all of the dark blues = ocean

```python
landPixels = []
for color in indexedColors:
    if color[1] in [0, 1, 2, 3]:
        landPixels.append(color[0])
landArea = sum(landPixels)
totalArea = sum(counts)
print("Fraction covered by land: " + str(landArea / totalArea))
```

```python
urbanPixels = []
for color in indexedColors:
    if color[1] == 2:
        urbanPixels.append(color[0])
urbanArea = sum(urbanPixels)
print("Fraction covered by urban area: " + str(urbanArea / totalArea))
```

```python
labeled_image = np.array(imageIndexed)
bw_urban = (labeled_image == 2)
plt.imshow(bw_urban, cmap = "gray")
plt.show()
```

### I found that urban areas are represented by the light blue, but there is a bright streak that is also light blue, which mistakenly overestimates the amount of urban area. This streak could be due to a fire causing a lot of smoke to rise.
