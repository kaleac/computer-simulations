```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Part 1

```python
ms = 1e7
mp = 1e4
ma = 1
G = 1
R = 20
```

```python
xs = 0
xp = 20
rs = np.array([xs, 0])
rp = np.array([xp, 0])
rCM = (rp*mp + rs*ms) / (mp+ms)
```

```python
# xs -= rCM 
# xp -= rCM
# rs = np.array([xs, 0])
# rp = np.array([xp, 0])
rs = rs - rCM
rp = rp - rCM
rCM = (rp*mp + rs*ms) / (mp+ms)

xa = R * np.cos(np.radians(45))
ya = R * np.sin(np.radians(45))
ra = np.array([xa, ya])
```

## Part 2

```python
G = 1
mus = (mp * ms) / (mp + ms)
mua = (ma * ms) / (ma + ms)
omegas = (G * mp * ms / (mus * R**3)) ** 0.5
omegaa = (G * ma * ms / (mua * R**3)) ** 0.5
vs = [0, omegas * rs[0]]
vp = [0, omegas * rp[0]]
va = [-omegas * ra[1], omegas * ra[0]]
y = np.concatenate([rs, vs, rp, vp, ra, va]).tolist()
```

## Part 3

```python
def KeplerODE(t, y):
    global mp, ms, ma, G
    
    rs = y[0: 2]
    vs = y[2: 4]
    rp = y[4: 6]
    vp = y[6: 8]
    ra = y[8: 10]
    va = y[10: 12]
    
    def F(m1, m2, r1, r2):
        r = np.array(r1) - np.array(r2)
#         print(r1)
#         print(r2)
        return -G * m1 * m2 * r / (np.linalg.norm(r)**3)
    drdts = vs
    drdtp = vp
    drdta = va
    
    dvdts = (F(ms, mp, rs, rp) + F(ms, ma, rs, ra)) / ms
    dvdtp = (F(mp, ms, rp, rs) + F(mp, ma, rp, ra)) / mp
    dvdta = (F(ma, ms, ra, rs) + F(ma, mp, ra, rp)) / ma
    
    return np.concatenate([drdts, dvdts, drdtp, dvdtp, drdta, dvdta])
```

## Part 4

```python
ycopy = y[:]
ycopy2 = y[:]
```

```python
P = np.sqrt(4 * (np.pi**2) * (R**3) / (G*(mp + ms)))
dt = P / 1000

xts = []
yts = []
xtp = []
ytp = []
xta = []
yta = []

t = 0
while t < P/4:
    rs = y[0: 2]
    vs = y[2: 4]
    rp = y[4: 6]
    vp = y[6: 8]
    ra = y[8: 10]
    va = y[10: 12]
    
    xts.append(rs[0])
    yts.append(rs[1])
    xtp.append(rp[0])
    ytp.append(rp[1])
    xta.append(ra[0])
    yta.append(ra[1])
    
    f1 = KeplerODE(t, y)
    f2 = KeplerODE(t+dt/2.0, y+f1*dt/2.0)
    f3 = KeplerODE(t+dt/2.0, y+f2*dt/2.0)
    f4 = KeplerODE(t+dt     ,y+f3*dt    )

    y = y + (f1 + 2.0*f2 + 2.0*f3 + f4) / 6.0 * dt
    t = t + dt
    
```

```python
plt.plot(xts, yts,'r-')
plt.plot(xtp, ytp, 'b-')
plt.plot(xta, yta, 'g-')

#plt.plot(0,0,'*',mfc='w',ms=10)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
def TransformIntoPlanetFrame(xp,yp,x,y):
    phi = np.arctan2(yp,xp)
    s = np.sin(phi)
    c = np.cos(phi)

    A = np.array( [ [c,s] , [-s,c] ] )

    r  = np.array([x,y])
    rt = A.dot(r) # matrix vector multiplication

    xt = rt[0]
    yt = rt[1]
    
    return [xt,yt] #return the transformed coordinates
```

```python
rp = y[4: 6]
vp = y[6: 8]
ra = y[8: 10]
va = y[10: 12]

TransformIntoPlanetFrame(rp[0], rp[1], ra[0], ra[1])
```

```python
xtaTransformed = []
ytaTransformed = []

y = ycopy[:]
t = 0
while t < (20 * P):
    rs = y[0: 2]
    vs = y[2: 4]
    rp = y[4: 6]
    vp = y[6: 8]
    ra = y[8: 10]
    va = y[10: 12]
    
    f1 = KeplerODE(t, y)
    f2 = KeplerODE(t+dt/2.0, y+f1*dt/2.0)
    f3 = KeplerODE(t+dt/2.0, y+f2*dt/2.0)
    f4 = KeplerODE(t+dt     ,y+f3*dt    )
    
    transformed = TransformIntoPlanetFrame(rp[0], rp[1], ra[0], ra[1])
#     y[10] = transformed[0]
#     y[11] = transformed[1]
    
    xtaTransformed.append(transformed[0])
    ytaTransformed.append(transformed[1])
    
    y = y + (f1 + 2.0*f2 + 2.0*f3 + f4) / 6.0 * dt
    t = t + dt
```
