```python
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import numpy as np
```

```python
ms = 3e+6
mp = 1
G  = 1e-6
```

```python
# initial positions
r = np.array([1, 0])

# initial velocities
v = np.array([0, 2]);
```

```python
t = 0
tMax = 10.0
dt   = 0.001;
n    = tMax / dt
```

```python
xt = []
yt = []
vxy = []
vyt = []
kint = []
pott = []
et = []
tt = []
```

```python
def KeplerODE(t,y):
    global mp,ms,G

    r = y[0:2]
    v = y[2:4]

    drdt = v 

    F    = - ms * mp * G / np.linalg.norm(r)**3 * r
    a    = F / mp
    dvdt = a
    
    return np.concatenate((drdt,dvdt))
```

```python
y = np.concatenate((r,v))
while (t<tMax):
    r = y[0:2]
    v = y[2:4]

    kin = 1.0/2.0*mp* np.linalg.norm(v)**2
    pot = -mp*ms*G / np.linalg.norm(r)
    e   = kin+pot
    #if (t==0.0):
    #    e0 = e
    #print(t,e,e0,e-e0)
    
    xt.append(r[0])
    yt.append(r[1])
    kint.append(kin)
    pott.append(pot)
    et.append(e)
    tt.append(t)
    
    #dydt = KeplerODE(t,y)
    #y = y + dydt*dt    

    f1 = KeplerODE(t       ,y          )
    f2 = KeplerODE(t+dt/2.0,y+f1*dt/2.0)
    f3 = KeplerODE(t+dt/2.0,y+f2*dt/2.0)
    f4 = KeplerODE(t+dt    ,y+f3*dt    )

    y = y + (f1 + 2.0*f2 + 2.0*f3 + f4) / 6.0 * dt
    t = t + dt
```

```python
plt.plot(xt,yt,'r-')
plt.plot(0,0,'*',mfc='w',ms=10)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
