```python
import obspy as op
import numpy as np
from obspy.clients.fdsn import Client

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import datetime
```

## Part 1

```python
currentDT = datetime.datetime.now()
print ('Download began: ',str(currentDT))

client = Client("IRIS")

# change the range of dates we look at 
t1 = op.UTCDateTime("1970-01-01T00:00:00") #start time of the request
t2 = op.UTCDateTime("2020-11-02T00:00:00") #end time of the request

# read https://docs.obspy.org/packages/autogen/obspy.clients.fdsn.client.Client.get_events.html
minLat = 33
maxLat = 37
minLon = -103
maxLon = -94
minMag = 5.0

catalog5 = client.get_events(starttime=t1, endtime=t2, minlatitude=minLat, maxlatitude=maxLat,
minlongitude=minLon, maxlongitude=maxLon, minmagnitude=minMag)

currentDT = datetime.datetime.now()
print ('Download ended: ',str(currentDT))
```

```python
print("Number of earthquakes from 1970 in Oklahoma: " + str(len(catalog5)))
```

```python
mag5s = []
for i in range(10):
    year = catalog5[i].origins[0].time.year
    mag5s.append(year)
```

```python
fig = plt.figure()
plt.hist(mag5s, bins = 50)
plt.show()
```

```python
catalog5.plot(projection='local')
plt.show()
```

## Part 3

```python
currentDT = datetime.datetime.now()
print ('Download began: ',str(currentDT))

client = Client("IRIS")

# change the range of dates we look at 
# t1 = op.UTCDateTime("2015-03-01T00:00:00") #start time of the request
# t2 = op.UTCDateTime("2015-04-01T00:00:00") #end time of the request
t1 = op.UTCDateTime("2019-07-01T00:00:00") #start time of the request
t2 = op.UTCDateTime("2019-09-01T00:00:00") #end time of the request

# read https://docs.obspy.org/packages/autogen/obspy.clients.fdsn.client.Client.get_events.html
minLat = 33
maxLat = 37
minLon = -103
maxLon = -94
minMag = 2.0 

catalog = client.get_events(starttime=t1, endtime=t2, minlatitude=minLat, maxlatitude=maxLat, minlongitude=minLon, maxlongitude=maxLon, minmagnitude=minMag)

currentDT = datetime.datetime.now()
print ('Download ended: ',str(currentDT))
```

```python
def dist(lat1, lon1, lat2, lon2):
    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)
    
    r = 6371
    
    dLat = r * (lat2 - lat1)
    dLon = (r * np.cos(lat2) * lon2) - (r * np.cos(lat1) * lon1)
    
    #a = np.sin(dLat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dLon/2)**2
    #c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    #dLat = R * (lat2 - lat1)
    #dLon = R * np.cos(lat2) * lon2 - R * np.cos(lat1) * lon1
    #return dLat ** 2 + dLon ** 2
    return np.sqrt(dLat**2 + dLon**2)
```

```python
count = 0
swarmLats = []
swarmLons = []

for i in range(len(catalog) - 1, -1, -1):
#     count = 0
#     swarmLats = []
#     swarmLons = []
    lat = catalog[i].origins[0].latitude    
    lon = catalog[i].origins[0].longitude    
    time = catalog[i].origins[0].time    
    timeDiffInSec = catalog[0].origins[0].time - catalog[i].origins[0].time
    for j in range(i-1, -1, -1):
        lat2 = catalog[j].origins[0].latitude    
        lon2 = catalog[j].origins[0].longitude    
        time2 = catalog[j].origins[0].time    
        
        dDist = dist(lat, lon, lat2, lon2)
        dTime = time2 - time
        if dDist < 10 and dTime < (14 * 60 * 60 * 24):
            swarmLats.append(lat2)
            swarmLons.append(lon2)
            count += 1
    if count > 40:
        break
```

```python
allLats = []
allLons = []
for event in catalog:
    lat = event.origins[0].latitude    
    lon = event.origins[0].longitude
    allLats.append(lat)
    allLons.append(lon)
```

```python
plt.scatter(allLons, allLats, color = "black")
plt.scatter(swarmLons, swarmLats, color = "blue")
plt.show()
```

## Bonus: Plot all swarms (I found 20)

```python
#count = 0
totalSwarms = 0
swarmLats = []
swarmLons = []

for i in range(len(catalog) - 1, -1, -1):
    count = 0
    currSwarmLats = []
    currSwarmLons = []
    lat = catalog[i].origins[0].latitude    
    lon = catalog[i].origins[0].longitude    
    time = catalog[i].origins[0].time    
    timeDiffInSec = catalog[0].origins[0].time - catalog[i].origins[0].time
    for j in range(i-1, -1, -1):
        lat2 = catalog[j].origins[0].latitude    
        lon2 = catalog[j].origins[0].longitude    
        time2 = catalog[j].origins[0].time    
        
        dDist = dist(lat, lon, lat2, lon2)
        dTime = time2 - time
        if dDist < 10 and dTime < (14 * 60 * 60 * 24):            
            currSwarmLats.append(lat2)
            currSwarmLons.append(lon2)
            count += 1
        
        if count >= 40:
            totalSwarms += 1
            swarmLats.extend(currSwarmLats)
            swarmLons.extend(currSwarmLons)
            break
            
print(totalSwarms)
```
